```python
Df=pd.DataFrame(data)
```

```python
#Define independent variables (X) and the dependent variable (y)
X = Df[['sqrt_ft', 'latitude', 'longitude']]
y=Df['sold_price']
```

```python
# Defining the split ratio
train_ratio = 0.8  # 80% train, 20% test
train_size = int(len(Df) * train_ratio)

# Splitting into training and testing sets
X_train = X[:train_size]
X_test = X[train_size:]
y_train = y[:train_size]
y_test = y[train_size:]

# Sizes of the train and test sets
print(f"Training set size: {X_train.shape[0]}")
print(f"Testing set size: {X_test.shape[0]}")
```

```python
#Normalize features (sqft, latitude, longitude)
def normalize(x):
    return (x - np.mean(x)) / np.std(x)

Df['sqft_normalized'] = normalize(Df['sqrt_ft'])
Df['latitude_normalized'] = normalize(Df['latitude'])
Df['longitude_normalized'] = normalize(Df['longitude'])
```

```python
# Prepare the feature matrix X (with bias term) and target vector y
X = np.ones((Df.shape[0], 4))  # Initialize X with a column of ones (bias term) and three features
X[:, 1] = Df['sqft_normalized']
X[:, 2] = Df['latitude_normalized']
X[:, 3] = Df['longitude_normalized']
```

```python
y = Df['sold_price'].values.reshape(-1, 1)
```

```python
#Linear Regression using Normal Equation

def normal_equation(X, y):
    return np.linalg.inv(X.T @ X) @ (X.T @ y)
```

```python
theta = normal_equation(X, y)
```

```python
# Step 4: Evaluate the model by calculating Mean Squared Error (MSE)
def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)
```

```python
# Prediction on the training data
y_pred = X @ theta
```

```python
y_pred
```

```python
# Calculate MSE
mse = mean_squared_error(y, y_pred)
```
